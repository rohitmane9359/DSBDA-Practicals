```python
 #that combines the values of several features in that row
def combine_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

```python
#filling in missing values, you are then applying the combine_features 
#function to each row of the dataframe df using the apply method
for feature in features:
    df[feature] = df[feature].fillna('')

df["combined_features"] = df.apply(combine_features,axis=1)
```
